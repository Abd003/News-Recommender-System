```python
dr1 = df[(df.category == "ENTERTAINMENT")]
dr2 = df[(df.category == "BUSINESS")]
dr3 = df[(df.category == "TECH")]
dr4 = df[(df.category == "SPORTS")]
dr5 = df[(df.category == "POLITICS")]
print(dr1.shape, dr2.shape, dr3.shape, dr4.shape, dr5.shape)
```

```python
df_row_reindex1 = pd.concat([dr1, dr2, dr3, dr4, dr5], ignore_index=True)
df_row_reindex1
```

```python
unique_links_tc = dr3['link'].tolist()
categories_tc = dr3['category'].tolist()
```

```python
unique_page_links1 = []
for link in unique_links_tc:
    link = 'https://www.huffpost.com'+link[30:]
    unique_page_links1.append(link)
print("No. of Unique Links : ", len(unique_page_links1))
print("Length of Category List : ", len(categories_tc))
```

```python
unique_articles = []
for i in range(len(unique_page_links1)):
    url = unique_page_links1[i]
    cat = categories_tc[i]
    article = Article(url)
    try:
        article.download()
        article.parse()
        article.nlp()
        print(url, cat, i)
        dr.loc[len(dr.index)] = [i, article.text, cat]
    except:
        pass
```

```python
dr.to_json('testfinal_tech3.json', orient='records', lines=True)
files.download('testfinal_tech3.json')
```
